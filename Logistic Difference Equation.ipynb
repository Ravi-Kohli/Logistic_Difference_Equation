# The Logistic Difference Equation

In this notebook, we will explore the properties of the $\textbf{logistic difference equation} $, which is given by:

$$ x_{n + 1} = r x_{n} \left( 1 - x_{n} \right) $$

where $ 0 \leq x \leq 1 $ and $ 0 \leq r \leq 4 $

```python
# importing the required modules
import numpy as np
import matplotlib.pyplot as plt
```

### Initial Plot

```python
N = 100                     # number of iterations                   # initial value of population
r1 = 2.5                    # intial value of r
x0 = 0.5
```

```python
def x_iteration(x_0, N, r):
    
    x = np.zeros(N)   # initialise the list to store the x values 
    x[0] = x_0
    
    for j in range(1,N):
        x[j] = r * x[j - 1] * (1 - x[j - 1])
    
    return x
```

```python
x1 = x_iteration(x0, N, r1)
```

```python
plt.plot(x1)
plt.grid()
plt.xlabel('Number of Iterations (N)')
plt.ylabel('X[n]')
plt.title('Plot 1: r = {} , $ x_0 $ = {}'.format(r1,x1[0]))
```

### Exploring the Effect of Varying r

Let us first do a few plots of the iterations for different values of r, using the same starting value of $ x_0 $

```python
r2 = 1
x2 = x_iteration(x0, N, r2)

r3 = 2.5
x3 = x_iteration(x0, N, r3)

r4 = 3
x4 = x_iteration(x0, N, r4)

r5 = 4
x5 = x_iteration(x0, N, r5)
```

```python
plt.subplot(2,2,1)
plt.plot(x2)
plt.grid()
plt.xlabel('Number of Iterations (N)')
plt.ylabel('X[n]')
plt.title('Plot 2: r = {} , $ x_0 $ = {}'.format(r2,x2[0]))

plt.subplot(2,2,2)
plt.plot(x3)
plt.grid()
plt.xlabel('Number of Iterations (N)')
plt.ylabel('X[n]')
plt.title('Plot 3: r = {} , $ x_0 $ = {}'.format(r3,x3[0]))

plt.subplot(2,2,3)
plt.plot(x4)
plt.grid()
plt.xlabel('Number of Iterations (N)')
plt.ylabel('X[n]')
plt.title('Plot 4: r = {} , $ x_0 $ = {}'.format(r4,x4[0]))

plt.subplot(2,2,4)
plt.plot(x5)
plt.grid()
plt.xlabel('Number of Iterations (N)')
plt.ylabel('X[n]')
plt.title('Plot 5: r = {} , $ x_0 $ = {}'.format(r5,x5[0]))
```

We can see that changing the value of r does change the 'steady state' value of the function. Let us now explicitly plot the value of r against the steady state value of the function using our $ x_{0} $

```python
r_list = np.linspace(0,4,4*N)
steady_state_values = np.zeros(4*N)
```

```python
for i, r in enumerate(r_list):
    
    # call the function
    y = x_iteration(x0, N, r)
    
    # store the steady state value
    steady_state_values[i] = y[-1]
```

```python
plt.plot(r_list, steady_state_values)
plt.grid()
plt.xlabel('Value of $ r $')
plt.ylabel('Steady State value')
```

### Exploring the Effect of Varying $x _{0} $

```python
x_list = np.linspace(0,1,2*N)
steady_state_values_x = np.zeros(2*N)
r_x = 2.5
```

```python
for i, x in enumerate(x_list):
    
    # call the function
    y = x_iteration(x, N, r_x)
    
    # store the steady state value
    steady_state_values_x[i] = y[-1]
```

```python
plt.plot(x_list, steady_state_values_x)
plt.grid()
plt.xlabel('Value of $ x_{0} $')
plt.ylabel('Steady State value')
```

We can see that we almost have a top-hat looking function. For $ x_{0} $ values at the extremeties, the value of the steady state drops to 0, but for all the values in-between, the steady state value is constant at $ \approx 0.6 $
